```python
X=data.drop(['Id','Species'],axis=1)
y = data.Species
```

```python
df = pd.DataFrame(X,columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```

```python
df.head()
```

```python
#we have to convert the species(String)to (numerical)using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
#to convert data into binary form using One-Hot-Encoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
encoder.fit_transform(y.reshape(-1,1))
```

```python
pd.scatter_matrix(df,c=y,figsize=[8,8],s=150,marker='D')
```

```python
#using knn,LogisticRegression,Kmean,DescisionTreeClassifier using pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
print(X.shape)
print(y.shape)
```

```python
#using knn
knn = KNeighborsClassifier()
knn.fit(X,y)
```

```python
print('Knn score:{}'.format(knn.score(X,y)))
```

```python
#using Logistic regression
log = LogisticRegression()
log.fit(X,y)
```

```python
print("logisticRegression score:{}".format(log.score(X,y)))
```

```python
#Descisiontreeclassifier
dt = DecisionTreeClassifier()
dt.fit(X,y)
```

```python
print("DecisionTree score:{}".format(dt.score(X,y)))
```

```python
#using ensamble method
from sklearn.ensemble import BaggingClassifier
bc = BaggingClassifier()
bc.fit(X,y)
```

```python
print("Bagging classifier score:{}".format(bc.score(X,y)))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(X,y)
```

```python
print("Adaboostclassifier score:{}".format(ada.score(X,y)))
```
