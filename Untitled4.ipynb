```python
def clear_content(text,remove_stopwords=True,lemmitization=True):
    text = text.lower()
    lemm = WordNetLemmatizer()
    if True : 
        text = text.split()
        new_text=[]
        for word in text:
            if word in contraction:
                new_text.append(contraction[word])
            else:
                new_text.append(word)
        text = " ".join(new_text)
    
    if remove_stopwords:
        text = text.split()
        stops = set(stopwords.words('english')+list(string.punctuation))
        text = [w for w in text if not w in stops]
        text = " ".join(text)
    if lemmitization:
        text = [lemm.lemmatize(w) for w in word_tokenize(text)]
        text = " ".join(text)
    return text
        
    
   
        
```

```python
clear_text=[]
for s in df.Course_Contents:
    clear_text.append(clear_content(s,remove_stopwords=True,lemmitization=True))
print("cleared the text")
```

```python
clear_text
```

```python
clean_content = []
for i in clear_text:
    clean_content.append(re.sub('\\b[A-Za-z]\\b|\\b[A-Za-z]\\b','',i))
    
```

```python
clean_content
```

```python
clear_course = []
for s in clean_content:
    clear_course.append(re.sub(r'[_"\--;%()|+&=*%.,!?:#$@\[\]/<>]','',s))
```

```python
clear_course
```

```python
import nltk
words = set(nltk.corpus.words.words())
clean=[]
for i in clear_course:
    clean.append(" ".join(w for w in nltk.wordpunct_tokenize(i) if w.lower() in words or not w.isalpha()))
```

```python
clean
```

```python
df['key_word'] = clean
```

```python
df.head()
```

```python
#Tf-idf vectorizer
tfidf = TfidfVectorizer()
count_matrix = tfidf.fit_transform(df['key_word'])
```

```python
count_matrix.shape
```

```python
#cosine simmilarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(count_matrix,count_matrix)
cos_sim
```

```python
#Recommendation
indices = pd.Series(df.index)
#defining the function that takes the title
#as the input and takes the top 10 recommendation
def recommendation(title,cos_sim = cos_sim):
    #definig the empty array
    recommend = []
    title_index = df[(df['Courses']==title)].index[0]
    #getting the index that matches with the title
    idx = indices[indices==title_index].index[0]
    #creating series with simmilarity score in descending order
    score_sim = pd.Series(cos_sim[idx]).sort_values(ascending=False)
    #getting the index of top 10 movies
    top_10_movie = list(score_sim.iloc[1:6].index)
    #[1:3] indicates that returns best matching of top 2 courses
                                                      
    #populating best 10 matches 
    for i in top_10_movie:
        recommend.append(list(df.index)[i])
    return recommend
    
```
